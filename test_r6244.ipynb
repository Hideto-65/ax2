```python
import pyvisa
import time
```

```python
# Resource Manager はvisaデバイスとの接続窓口
rm = pyvisa.ResourceManager()

print("検出されたvisaリソース一覧")
for res in rm.list_resources():
    print(" ", res)
```

```python
Adress_6244 = "GPIB0::8::INSTR"

# resource　visa coneccitng
src = rm.open_resource(Adress_6244)

print("6221 IDN", src.query("*IDN?").strip())
```

```python
src.write("*RST")          # リセット
```

```python
src.write("*CLS")          # ステータスクリア
```

```python
src.write("IF")          # current generation mode
# not needed
```

```python
src.write("MD0")      # dc mode
```

```python
#src.write("DV")      # current generation mode
```

```python
src.query("R?") #auto range mode?
```

```python
src.write(f"D100UA") # current setting,limit voltage setting
```

```python
src.write("I0")      # current range 
# can not activate
```

```python
src.query("I?")
```

```python
src.write("H") #output pff
```

```python
# 設定したい電流値 [A]
TARGET_CURRENT = 1e-6   # 1 µA
src.write(f":SOUR:CURR {TARGET_CURRENT}")
```

```python
src.write("E") #output_on
```
