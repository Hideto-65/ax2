```python
import serial
import serial.tools.list_ports
import time
```

```python
# port 一覧を表示
ports = serial.tools.list_ports.comports()
for p in ports:
    print(p.device, p.description)
```

```python
ser = serial.Serial(
    port="COM6",                      # シリアルポート名 (Windows: COM1, COM2..., Mac/Linux: /dev/ttyUSB0 など)
    baudrate=9600,                    # 通信速度 (bps)
    bytesize=serial.EIGHTBITS,        # データビット長 (5,6,7,8)
    parity=serial.PARITY_NONE,        # パリティビット (NONE, EVEN, ODD, MARK, SPACE)
    stopbits=serial.STOPBITS_ONE,     # ストップビット (1, 1.5, 2)
    timeout=1                         # 読み込みタイムアウト (秒)
    rtscts=True,          # RTS/CTS を使う
)
```

```python
# --- 2. 装置にコマンドを送る関数（便利） ---
def send(cmd):
    """GSC-02 にコマンドを送り、応答を返す"""
    full_cmd = (cmd + "\r\n").encode() #コマンドの終端を示し、バイト列に変換
    ser.write(full_cmd) # コマンド送信
    time.sleep(0.1)
    resp = ser.readline().decode().strip() # コマンドの読み込み
    print(f"CMD: {cmd}, RESP: {resp}")
    return resp
```

```python
# --- 3. 原点復帰（安全のため必須） ---
send("H:1+")
```

```python
# 例：相対で 100 パルスだけ動かすという設定
send("M:1+1000")
```

```python
# 実際の移動開始
send("G")
```

```python
# --- 5. 動作確認後にクローズ ---
ser.close()
```
